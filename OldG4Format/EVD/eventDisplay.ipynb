# Event Display: Supernova Neutrino + Radilogicals

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
detector_bounds = [[-150, 150], [895.94, 1195.94], [-325, 325]] # cm (y, z, x)
def draw_boundaries(ax):
    """
    Draw the detector boundaries as a wireframe
    not needed, but pretty
    """
    boundKwargs = {'color': 'black',
                   'ls': '--'}
    
    ax.plot([detector_bounds[0][0], detector_bounds[0][1]],
            [detector_bounds[1][0], detector_bounds[1][0]],
            [detector_bounds[2][0], detector_bounds[2][0]],
            **boundKwargs)
    ax.plot([detector_bounds[0][0], detector_bounds[0][1]],
            [detector_bounds[1][1], detector_bounds[1][1]],
            [detector_bounds[2][0], detector_bounds[2][0]],
            **boundKwargs)
    ax.plot([detector_bounds[0][0], detector_bounds[0][1]],
            [detector_bounds[1][0], detector_bounds[1][0]],
            [detector_bounds[2][1], detector_bounds[2][1]],
            **boundKwargs)
    ax.plot([detector_bounds[0][0], detector_bounds[0][1]],
            [detector_bounds[1][1], detector_bounds[1][1]],
            [detector_bounds[2][1], detector_bounds[2][1]],
            **boundKwargs)

    ax.plot([detector_bounds[0][0], detector_bounds[0][0]],
            [detector_bounds[1][0], detector_bounds[1][1]],
            [detector_bounds[2][0], detector_bounds[2][0]],
            **boundKwargs)
    ax.plot([detector_bounds[0][1], detector_bounds[0][1]],
            [detector_bounds[1][0], detector_bounds[1][1]],
            [detector_bounds[2][0], detector_bounds[2][0]],
            **boundKwargs)
    ax.plot([detector_bounds[0][0], detector_bounds[0][0]],
            [detector_bounds[1][0], detector_bounds[1][1]],
            [detector_bounds[2][1], detector_bounds[2][1]],
            **boundKwargs)
    ax.plot([detector_bounds[0][1], detector_bounds[0][1]],
            [detector_bounds[1][0], detector_bounds[1][1]],
            [detector_bounds[2][1], detector_bounds[2][1]],
            **boundKwargs)

    ax.plot([detector_bounds[0][0], detector_bounds[0][0]],
            [detector_bounds[1][0], detector_bounds[1][0]],
            [detector_bounds[2][0], detector_bounds[2][1]],
            **boundKwargs)
    ax.plot([detector_bounds[0][0], detector_bounds[0][0]],
            [detector_bounds[1][1], detector_bounds[1][1]],
            [detector_bounds[2][0], detector_bounds[2][1]],
            **boundKwargs)
    ax.plot([detector_bounds[0][1], detector_bounds[0][1]],
            [detector_bounds[1][0], detector_bounds[1][0]],
            [detector_bounds[2][0], detector_bounds[2][1]],
            **boundKwargs)
    ax.plot([detector_bounds[0][1], detector_bounds[0][1]],
            [detector_bounds[1][1], detector_bounds[1][1]],
            [detector_bounds[2][0], detector_bounds[2][1]],
            **boundKwargs)

    return ax
```

```python
SNFile = '/Users/yuntse/data/lartpc_rd/gampix/OldFormat/g4/mixed/rad_nueArCC_garching_nh_mxDir_xpos-105cm_g4_0000.csv'
```

```python
SNEvent = pd.read_csv( SNFile )
SNEvent
```

Pick an event

```python
iEvt = 0
```

```python
Evt = SNEvent[(SNEvent.event==iEvt)]
Evt
```

Select the segments which deposit energy for signals and radiologicals

```python
sig = Evt[(Evt.isSignal==1)&(Evt.pdgCode!=12)&(Evt.dE>0.)]
```

For the radiologicals, we will see the ones drifting to the CRP at the time window of the neutrino interactions $\pm$ 100 cm.  
In this case, the interaction happens at t = 0ns, x = 0cm.  
The signal at x = 0cm will arrive at the CRP at t = 2.03ms.  
The signal at x = 150cm will arrive at the CRP at t = 1093750ns, and similarly, the signal at x = -150cm will arrive at the CRP at t = 2968750ns.
Also select the deposit energy within -3250mm (cathode) < startX < 3250mm (CRP)

```python
rad = Evt[(Evt.isSignal==0)&(Evt.dE>0.)&(Evt.startT+(3250-Evt.startX)/1.6*1000>1093750)&(Evt.startT+(3250-Evt.startX)/1.6*1000<2968750)&(Evt.startX>-3250.)&(Evt.startX<3250.)]
```

```python
fig = plt.figure( figsize = (9, 9))  
ax = fig.add_subplot( 1, 1, 1, projection = '3d' )

ax.scatter( sig['startY']/10, sig['startZ']/10, sig['startX']/10, label = 'Supernova neutrino')
ax.scatter( rad['startY']/10, rad['startZ']/10, rad['startX']/10, c = 'gold', label = 'Radiologicals')
                    
ax.scatter( 0., 1045.94, 0., s = 800, c = '#DC143C', label = 'Interaction vertex', marker = '*' )
draw_boundaries(ax)
ax.set_xlabel('y [cm]', fontsize = 18, labelpad = 10)
ax.set_ylabel('z [cm]', fontsize = 18, labelpad = 20)
ax.set_zlabel('x [cm]', fontsize = 18, labelpad = 10)
ax.tick_params( axis = 'x', labelsize = 18 )
ax.tick_params( axis = 'y', labelsize = 18 )
ax.tick_params( axis = 'z', labelsize = 18 )
ax.legend( ncol = 2, fontsize = 14, loc = 'upper left' )
ax.set_xlim( [-160, 160])
ax.set_ylim( [885, 1205])
ax.set_zlim( [-335, 335])
# fig.subplots_adjust( right = 0.9)

fig.savefig('SNRad.pdf')
fig.savefig('SNRad.png')
```
