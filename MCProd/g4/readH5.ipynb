Pandas version 2 will not implicitly accept columns containing arrays (like the vectors in the `trajectories` data).
We need to convert explicitly all the columns that have an array structure (`ndim > 1`) while we can use directly all the others:

```python
dT = pd.DataFrame({
    name: tempT[name].tolist() if tempT[name].ndim > 1 else tempT[name]
    for name in tempT.dtype.names
})
```
