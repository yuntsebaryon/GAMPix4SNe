# Garching Supernova Neutrino Flux

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
params = pd.read_csv('/Users/yuntse/data/supernova/garching/garching_pinched_info.dat', header = None, sep = r'\s+', 
                        names = ['No', 'nue_alpha', 'nuebar_alpha', 'nux_alpha', 'nue_E', 'nuebar_E', 'nux_E', 'nue_L', 'nuebar_L', 'nuxL' ])
```

```python
params
```

```python
t = pd.read_csv('/Users/yuntse/data/supernova/garching/garching_pinched_info_key.dat', header = None, sep = r'\s+', 
                names = ['No', 't', 'dt'])
t
```

```python
fig, ax = plt.subplots( figsize = (8, 4), layout = 'tight' )

ax.plot( t['t'], params['nue_L'], lw = 2)
ax.set_xlabel( r't [second]', fontsize = 16 )
ax.set_ylabel( r'Luminosity', fontsize = 16 )
ax.tick_params(axis = 'x', labelsize = 16)
ax.tick_params(axis = 'y', labelsize = 16)
ax.yaxis.get_offset_text().set_fontsize(16)
plt.xscale('log')
plt.yscale('log')
#ax.set_xlim( [0, 0.05])

ax.grid(":")
```

```python
fig, ax = plt.subplots( figsize = (8, 4), layout = 'tight' )

ax.plot( t['t'], t['dt'], lw = 2)
ax.set_xlabel( r't [second]', fontsize = 16 )
ax.set_ylabel( r'dt', fontsize = 16 )
ax.tick_params(axis = 'x', labelsize = 16)
ax.tick_params(axis = 'y', labelsize = 16)
ax.yaxis.get_offset_text().set_fontsize(16)
plt.xscale('log')

ax.grid(":")
```

```python
fig, ax = plt.subplots( figsize = (8, 4), layout = 'tight' )

ax.plot( t['t'], params['nue_L']/t['dt'], lw = 2)
ax.set_xlabel( r't [second]', fontsize = 16 )
ax.set_ylabel( r'Luminosity', fontsize = 16 )
ax.tick_params(axis = 'x', labelsize = 16)
ax.tick_params(axis = 'y', labelsize = 16)
ax.yaxis.get_offset_text().set_fontsize(16)
plt.xscale('log')
plt.yscale('log')

ax.grid(":")
```
