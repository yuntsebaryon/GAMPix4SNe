```python
# create a dataframe for fixing the color code
pdgCodes = dT.pdg_id.unique()

particleName = np.array([ 'electron', 'photon', 'alpha', 'positron', 'neutron', 
                          'Ar41', 'Ar40', 'Ca43',
                         ])

pdgDF = pd.DataFrame({ 'pdgCode': pdgCodes, 'name': particleName })

pdgDF
```

```python
palette = np.vstack([
    plt.get_cmap("tab20").colors,
])[:8]

# build mapping using pdgDF order
code_to_color = dict(zip(pdgDF["pdgCode"], palette))

# add to pdgDF
pdgDF["color"] = pdgDF["pdgCode"].map(code_to_color)

pdgDF
```

```python
fig, ax = plt.subplots()

ipdg = 0
ax.hist(dT.loc[dT['pdg_id'] == pdgDF['pdgCode'].iloc[ipdg], 'E_start']-0.511, bins = np.linspace(0., 3.5, 71), 
        histtype = 'step', color = pdgDF['color'].iloc[ipdg], label = pdgDF['name'].iloc[ipdg])
ax.set_xlabel(r'Kinetic Energy [MeV]')
ax.set_xlim( [0, 3.5] )
ax.set_yscale('log')
ax.legend()
```
